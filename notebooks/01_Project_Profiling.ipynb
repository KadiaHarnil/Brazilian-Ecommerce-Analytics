# 5. Dataset Overview

## Customers Dataset

- One row represents one customer.
- Primary Key: customer_id
- Contains customer location information.
- Useful for customer and geographical analysis.

## Payments Dataset

- **One row represents:** One payment transaction made for an order.
- **Primary Key:** Combination of `order_id` and `payment_sequential`
- **Foreign Key:** `order_id`
- **Contains:** Payment details such as payment method, payment value, payment installments, and payment sequence.
- **Useful for:** Revenue analysis, payment method analysis, installment behavior, and customer payment trends.

---

## Order Items Dataset

- **One row represents:** One product item purchased in an order.
- **Primary Key:** Combination of `order_id` and `order_item_id`
- **Foreign Keys:** `order_id`, `product_id`, `seller_id`
- **Contains:** Product price, freight value, seller information, shipping limit date, and purchased product details.
- **Useful for:** Sales analysis, revenue analysis, seller performance, product performance, and shipping cost analysis.

---

## Products Dataset

- **One row represents:** One product available on the e-commerce platform.
- **Primary Key:** `product_id`
- **Foreign Key:** None
- **Contains:** Product category, weight, dimensions, photos, and product description details.
- **Useful for:** Product category analysis, inventory insights, and product performance evaluation.

---

## Reviews Dataset

- **One row represents:** One customer review submitted for an order.
- **Primary Key:** `review_id`
- **Foreign Key:** `order_id`
- **Contains:** Review score, review comments, review creation date, and review response timestamp.
- **Useful for:** Customer satisfaction analysis, review trends, sentiment analysis, and evaluating the impact of delivery performance on customer experience.

---

## Sellers Dataset

- **One row represents:** One seller registered on the e-commerce platform.
- **Primary Key:** `seller_id`
- **Foreign Key:** None
- **Contains:** Seller location details such as city, state, and zip code.
- **Useful for:** Seller performance analysis, regional seller distribution, logistics analysis, and marketplace performance evaluation.

---

## Product Category Translation Dataset

- **One row represents:** One product category and its English translation.
- **Primary Key:** `product_category_name`
- **Foreign Key:** None
- **Contains:** Portuguese product category names and their English translations.
- **Useful for:** Improving data readability, reporting, dashboard creation, and category-based business analysis.

# Dataset Relationship

| Parent Table | Child Table | Key         | Relationship                                        |
| ------------ | ----------- | ----------- | --------------------------------------------------- |
| Customers    | Orders      | customer_id | One Customer → Many Orders                          |
| Orders       | Payments    | order_id    | One Order → Many Payments                           |
| Orders       | Reviews     | order_id    | One Order → One/Many Reviews                        |
| Orders       | Order Items | order_id    | One Order → Many Items                              |
| Products     | Order Items | product_id  | One Product → Many Order Items                      |
| Sellers      | Order Items | seller_id   | One Seller → Many Order Items                       |
