## Data Cleaning & Quality Assessment

This notebook evaluates the quality of the raw datasets before analysis. It identifies missing values, duplicate records, incorrect data types, and other potential data quality issues. The datasets are then cleaned and prepared for exploratory data analysis (EDA).

```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("../data/olist_customers_dataset.csv")
orders = pd.read_csv("../data/olist_orders_dataset.csv")
payments = pd.read_csv("../data/olist_order_payments_dataset.csv")
order_items = pd.read_csv("../data/olist_order_items_dataset.csv")
products = pd.read_csv("../data/olist_products_dataset.csv")
reviews = pd.read_csv("../data/olist_order_reviews_dataset.csv")
sellers = pd.read_csv("../data/olist_sellers_dataset.csv")
translation = pd.read_csv("../data/product_category_name_translation.csv")
```

```python
summary = pd.DataFrame({
    'Dataset':[
        "Customers",
        "Orders",
        "Order Payment",
        "Order Items",
        "Products",
        "Order Reviews",
        "Sellers",
        "Category Translation"
    ],
    "Rows": [
        customers.shape[0],
        orders.shape[0],
        payments.shape[0],
        order_items.shape[0],
        products.shape[0],
        reviews.shape[0],
        sellers.shape[0],
        translation.shape[0]
    ],
    "Columns": [
        customers.shape[1],
        orders.shape[1],
        payments.shape[1],
        order_items.shape[1],
        products.shape[1],
        reviews.shape[1],
        sellers.shape[1],
        translation.shape[1]
    ],
    "Missing Value":[
        customers.isnull().sum().sum(),
        orders.isnull().sum().sum(),
        payments.isnull().sum().sum(),
        order_items.isnull().sum().sum(),
        products.isnull().sum().sum(),
        reviews.isnull().sum().sum(),
        sellers.isnull().sum().sum(),
        translation.isnull().sum().sum(),
    ],
    "Duplicate Value":[
        customers.duplicated().sum().sum(),
        orders.duplicated().sum().sum(),
        payments.duplicated().sum().sum(),
        order_items.duplicated().sum().sum(),
        products.duplicated().sum().sum(),
        reviews.duplicated().sum().sum(),
        sellers.duplicated().sum().sum(),
        translation.duplicated().sum().sum(),
    ]
})
summary
```

## Initial Data Quality Summary

### Observations

- The Customers dataset contains no missing values or duplicate records.
- The Products dataset contains missing values that require further investigation.
- The Orders dataset includes timestamp columns that will need conversion to datetime.
- No duplicate rows were identified in most datasets.
- Overall, the datasets are in good condition, but some preprocessing is required before analysis.

```python
customers.info()
orders.info()
payments.info()
order_items.info()
products.info()
reviews.info()
sellers.info()
translation.info()
```

```python
orders.order_purchase_timestamp = pd.to_datetime(orders["order_purchase_timestamp"])
orders.order_delivered_carrier_date = pd.to_datetime(orders["order_delivered_carrier_date"])
orders.order_delivered_customer_date = pd.to_datetime(orders["order_delivered_customer_date"])
orders.order_estimated_delivery_date = pd.to_datetime(orders["order_estimated_delivery_date"])
orders.order_approved_at = pd.to_datetime(orders["order_approved_at"])
orders.info()
```
